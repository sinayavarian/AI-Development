```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

```python
# Calculate and report the mean, std, and var of the data shown above.
print("hours mean: ", no_hours.mean())
print("hours std: ", no_hours.std())
# What is the theoretical relationship between std and var? Do the numbers reperted in the previous step agree with your expectation?
#The theoretical relationship between variance and
#standard deviation is that the standard deviation is the square root of the variance.
#In this simulation, the expected standard deviation is 10 and the expected variance is 100.
#The observed values (std ≈ 9.98 and var ≈ 99.6) are very close to these theoretical values.
#The small difference is due to random sampling variability, so the results agree well with theoretical expectations.
```

```python
# Begin changing the different parameters on the normal function, discuss your observations as you change each input parameter.
np.random.seed(10)
no_hours = np.random.normal(40.0, 5.0, 10000)

plt.hist(no_hours, 50)
plt.show()
```

```python
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 100)

plt.hist(no_hours, 50)
plt.show()
```

```python
np.random.seed(10)
no_hours = np.random.normal(50.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()
#Changing the mean shifts the center of the distribution without affecting its spread.
```

```python
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 50000)

plt.hist(no_hours, 50)
plt.show()
#Increasing the sample size makes the distribution smoother and closer to the theoretical normal distribution.
```

```python
#Changing the mean shifts the center of the distribution without affecting its spread.
#Changing the standard deviation affects how spread out the data are around the mean, with larger values producing wider distributions.
#Increasing the sample size does not change the shape of the distribution but makes the results more stable and closer to the theoretical values.
```
